```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
houer=np.array([1,2,3,5,6,8,],dtype=float)
result=np.array([0,0,0,1,1,1],dtype=float)
print("the houers studies :",houer)
print("the result :",result)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(houer,result)
plt.xlabel("houer studies")
plt.ylabel("result =0 fail,result=1 pass")
plt.title("houer vs result")
plt.grid(True)
plt.show()
```

```python
def linear_neros(x,w,b):
  z=w*x+b
  return z

w=1.2
b=-4
z=linear_neros(houer,w,b)
print("the value z",z)
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
propapilty=sigmoid(z)
print("the propapilty is :")
print(np.round(propapilty,3))
```

```python
predict_class=(propapilty>=0.5).astype(int)
print("the predict class is :",result.astype(int))
print("propapilty is :",predict_class)
```

```python
x_line=np.linspace(0,10,100)
y_line=linear_neros(x_line,w,b)
z_line=sigmoid(x_line)
plt.figure(figsize=(10,5))
plt.scatter(houer,result,label="the acyual ")
plt.plot(x_line,y_line,label="the nerun curve ")
plt.axhline(0.5, linestyle="--", label="Decision Threshold = 0.5")
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")
plt.title("Prediction Using One Artificial Neuron")
plt.yticks([0, 0.5, 1], ["Fail", "0.5", "Pass"])
plt.legend()
plt.grid(True)
plt.show()
```
